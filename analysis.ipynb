```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import joblib

# Load the stock data
df = pd.read_csv("MBG.DE.csv")

# Select numeric features for clustering
features = ['Open', 'High', 'Low', 'Close', 'Volume']
X = df[features]

# Standardize the data (Essential for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create 3 distinct clusters (e.g., Low, Medium, High activity days)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_scaled)

# Save the model and scaler
joblib.dump(kmeans, "mbg_kmeans_model.pkl")
joblib.dump(scaler, "mbg_scaler.pkl")

print("Clustering assets saved!")
```
